# Calcul de DSL


The DSL is derived from ZOS by removing its global mean.

Donc :
DSL(x,y,t) = zos(x,y,t) - mean_WZ(t)


```python
def dsl(i):
    dsl=[]
    for t in range(210):
        dsl.append(ZOS[i][t,:,:] - mean_WZ[i][t])
    return dsl
```

```python
DSL=[]

for i in range(19):
    DSL.append(dsl(i))
    
```

GMTSL corresponds to DSL plus global mean thermosteric sea level

Donc :
GMTSL(x,y,t) = DSL(x,y,t) + zostoga(t)

```python
def ToE_GMTSL_both(i):
    ToE_GMTSL_pos = GMTSL_piC[0].copy()
    ToE_GMTSL_neg = GMTSL_piC[0].copy()
    
    for y in range(332):
        print(y)
        for x in range(362):
            
            P=0
            N=0
            Ap=1000
            An=1000
            
            for t in range(0,80):
                if (float(GMTSL_sur_SD_piC[i][t][y,x]))>=2:
                    P=P+1
                    N=0
                if (float(GMTSL_sur_SD_piC[i][t][y,x]))<=-2:
                    N=N+1
                    P=0
                else :
                    P=0
                    N=0
                if P==1 :
                    Ap = t
                if N==1 :
                    An = t

            if P>=20 and Ap<=60:
                ToE_GMTSL_pos[y,x]=1980+Ap
                ToE_GMTSL_neg[y,x]=np.nan
            if N>=20 and An<=60:
                ToE_GMTSL_neg[y,x]=1980+An
                ToE_GMTSL_pos[y,x]=np.nan
            else :
                ToE_GMTSL_neg[y,x]=np.nan 
                ToE_GMTSL_pos[y,x]=np.nan
                

            
    return ToE_GMTSL_pos,ToE_GMTSL_neg
                
#Ne pas oublier que si i>60, on peut pas conclure
```
